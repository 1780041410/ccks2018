```python
import pymysql
```

# 设置读写引擎

```python
read_con = pymysql.connect(host="192.168.1.13",user='root', password='root', database='ccks',charset='utf8')
```

```python
from sqlalchemy import create_engine
save_con = create_engine('mysql+pymysql://root:root@192.168.1.13:3306/ccks?charset=utf8')
```

```python
cur = read_con.cursor()
```

```python
import pandas as pd
import gc
```

# 导入pkubase

```python
sql = """CREATE TABLE `pkubase` (
	`id` INT(11) NOT NULL AUTO_INCREMENT,
	`entry` VARCHAR(80) NOT NULL,
	`prop` VARCHAR(50) NOT NULL,
	`value` VARCHAR(200) NOT NULL,
	PRIMARY KEY (`id`),
	INDEX `entry` (`entry`),
	INDEX `prop` (`prop`),
	INDEX `value` (`value`))
COLLATE='utf8_general_ci'
ENGINE=MyISAM
AUTO_INCREMENT= 0
"""
```

```python
# 创建pkubase
cur.execute(sql)
```

```python
with open('pkubase-triples.txt','rb') as f:
    while True:
        ans = []
        for i in range(10000):
            line = f.readline()  # 逐行读取
            if not line:  # 到 EOF，返回空字符串，则终止循环
                break
            ans.append(line.decode("utf8")[:-1].split("\t"))

        df  = pd.DataFrame(ans,columns=["entry","prop","value"])
        df.to_sql(name="pkubase",con = save_con ,if_exists="append",chunksize=5000,index=False)
        del df,ans
        gc.collect()
```
